```python
import sys, subprocess
def pipi(pkgs): subprocess.check_call([sys.executable, "-m", "pip", "install", *pkgs])

pipi([
    "geopandas", "pyogrio", "osmnx", "rasterio", "pysal", "splot",
    "scikit-gstat", "pykrige", "sqlalchemy", "psycopg[binary]",
    "folium", "plotly", "streamlit", "streamlit-folium"
])
print("Librerías geoespaciales instaladas correctamente.")
```
